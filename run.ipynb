# Generate image

```python
import random
import sys
sys.path.insert(0, './scripts')
from txt2img import main



prompt = 'ghouls coming out of a gas station in a post apocalyptic world'
num_samples = 1
n_iter = 5
seed = random.randint(sys.maxsize-1, sys.maxsize)
settings = ['--prompt', f"{prompt}", '--plms', '--n_iter', str(n_iter), '--n_samples', str(num_samples), '--seed', str(seed)]
print(settings)

main(settings)
```
